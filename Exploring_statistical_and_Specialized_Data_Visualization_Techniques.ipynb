```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Values':[0,70,50,80,90],
    'Sales':np.random.randint(100,500,5),
    'Profit':np.random.randint(10,100,5)
})
df.head()
```

```python
#1.Area Plot
plt.figure()
plt.fill_between(df['Category'],df['Values'],color='red',alpha=0.50)
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Area Plot')
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize= (7, 4))
plt.fill_between(df['Category'], df['Sales'], color = 'black', alpha = 1, label = 'Sales')
plt.fill_between(df['Category'], df['Profit'], color = 'grey', alpha = 1, label = 'Profit')
plt.title("Sales and profit area plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
```

```python
#2. Pie Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# To remove a segment from a pie chart, you need to filter your data.
# For example, let's remove Category 'A' from the pie chart:
filtered_df = df[df['Category'] != 'A']

plt.figure()
plt.pie(filtered_df['Values'], labels=filtered_df['Category'], autopct='%1.2f%%')
plt.title('Pie Chart (Category A removed)')
plt.show()
```

```python
#Explode only one slice (e.g.Python)
explode_values = [0.3, 0, 0, 0.1, 0] # Corresponds to A, B, C, D, E

plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%', explode=explode_values)
plt.title('Pie Chart with Exploded Slice (Category D)')
plt.show()
```

```python
#3 Donut Chart
plt.figure()
plt.pie(df['Values'],labels=df['Category'],autopct='%.1f%%')
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

gcf=Get current figure gca=Get Current Axes

```python
#4. Boxplot (Outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Boxplot of Sales")
plt.show()

```

```python
#5. Heatmap (Correlation)
plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
corr = df[['Values','Sales','Profit']].corr()

plt.figure(figsize=(5,4))

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Bubble Plot


plt.scatter(df['Sales'], df['Profit'],
            s=df['Values']*10)   # size of bubbles

plt.title("Bubble Plot")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(50, 300),      # min-max size
                palette='viridis')

plt.title("Bubble Plot using Seaborn")

plt.show()
```

```python
#2. Second Dataset
np.random.seed(7)
n = 100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df['Loan_Status'] = np.where(df['Credit_Score'] > 600, 'Approved', 'Rejected')

print(df.head())
```

```python
#Boxplot (Outlier Detection – Loan Amount)
import matplotlib.pyplot as plt

plt.boxplot(df['Loan_Amount'])

plt.title("Loan Amount Distribution (Boxplot)")
plt.ylabel("Loan Amount")

plt.show()
```

```python
#6. Correlation Heatmap (Very Important)
import seaborn as sns

corr = df[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Matrix")

plt.show()
```
